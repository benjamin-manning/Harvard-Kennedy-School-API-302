```python
%matplotlib inline
import math
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor

import warnings
```

```python
norway_df = pd.read_csv('Norway.csv')
sweden_df = pd.read_csv('Sweden.csv')
finland_df = pd.read_csv('Finland.csv')

norway_df['Death Rate'] = norway_df['Total Deaths']/norway_df['Total Cases']
sweden_df['Death Rate'] = sweden_df['Total Deaths']/sweden_df['Total Cases']
finland_df['Death Rate'] = finland_df['Total Deaths']/finland_df['Total Cases']
```

```python
frames = [norway_df, sweden_df, finland_df]
scan = pd.concat(frames)
scan.sort_values(by='Days')
```

```python
plt.plot(norway_df['Days'], norway_df['Cases Per Day'], label = 'norway')
plt.plot(sweden_df['Days'], sweden_df['Death Rate'], label = 'sweden')
plt.plot(finland_df['Days'],finland_df['Death Rate'], label = 'finland')
plt.xlabel('Days')
plt.ylabel('Death Rate')
plt.title("Death rate as compared to timeline in Days of the Pandemic (starting 1/22/2020)")
plt.legend()
```

```python
plt.scatter(norway_df['Deaths Per Day'], norway_df['Death Rate'], label = 'norway', alpha = .5)
plt.scatter(sweden_df['Deaths Per Day'], sweden_df['Death Rate'], label = 'sweden', alpha = .5)
plt.scatter(finland_df['Deaths Per Day'],finland_df['Death Rate'], label = 'finland', alpha = .5)
plt.title('Death rate as a function of Deaths per day')
plt.xlabel("Deaths Per Day")
plt.ylabel('Death Rate')
plt.legend()
```

```python
N_x = norway_df[['Days']]
N_y = norway_df['Cases Per Day']
model = KNeighborsRegressor(n_neighbors=10)
model.fit(N_x, N_y)

N_pred = model.predict(N_x)
plt.scatter(N_x, N_y, label='Cases Per day', color = 'r')
plt.plot(N_x, N_pred,label = 'Knn = 10 fit')
plt.title('Cases per day as a function of time in Norway')
plt.xlabel('Days')
plt.ylabel('Cases Per day')
plt.ylim(0,1750)
plt.legend()
```
